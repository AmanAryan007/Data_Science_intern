```python
import pandas as pd

def resample_data(data, timeframe):
    # Convert the 'Date' column to datetime format
    data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
    
    # Set the 'Date' column as the index
    data.set_index('Date', inplace=True)
  
   
    
    # Resample the data to the specified timeframe
    resampled_data = data.resample(timeframe).agg({'Open': 'first','High': 'max','Low': 'min','Close': 'last','Volume': 'sum','Time': 'first','Date_time': 'first'})
    resampled_data.dropna(inplace=True) 
    return resampled_data
# Load the Bank Nifty spot data from a CSV file
df = pd.read_csv('/content/drive/MyDrive/dataaaa/BN_Spot.csv')
# Drop rows with NaN values
df.dropna(inplace=True)

# Specify the desired timeframe
timeframe = '5T'  # Change this to '15T', '1H', '1D', or '1W' for other timeframes

# Resample the data based on the specified timeframe
resampled_df = resample_data(df, timeframe)
# Filter the resampled_df to include only rows between 9:15 and 15:30
selected_data = resampled_df[(resampled_df['Time'] >= '09:15') & (resampled_df['Time'] <= '15:30')]
```

```python
# Filter the resampled_df to include only rows between 9:15 and 15:30
selected_data = resampled_df[(resampled_df['Date_time'] >= '09:15') & (resampled_df['Date_time'] <= '15:30')]
```
