```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
import requests
```

```python
df= pd.read_excel("/content/Input (1) (1).xlsx", index_col=0)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import urllib.error

for i in range(0, len(df["URL"])):
    try:
        html = urlopen(df["URL"].iloc[i])
        bs = BeautifulSoup(html, "html.parser")
        titles = bs.find_all(['h1'])
        print('List all the header tags :', *titles, sep='\n\n')
    except urllib.error.HTTPError as e:
        print(f"error accessing URL: {df['URL'].iloc[i]} - {e}")
```

```python

for i in range(0,len(df["URL"])):
  response = requests.get(df["URL"].iloc[0])  # Replace with the URL of the website
  html_content = response.content

  # Create a BeautifulSoup object
  soup = BeautifulSoup(html_content, 'html.parser')

  div_element = soup.find('div', class_='td_block_wrap tdb_single_content tdi_130 td-pb-border-top td_block_template_1 td-post-content tagdiv-type')  # Replace 'your-class-name' with the actual class name

  # Find all <p> tags within the specific <div> element and extract the text
  p_tags = div_element.find_all('p')

  file_name = f'article_{i}.txt'
  file = open(file_name, 'w')

    # Write the content of each <p> tag to the file
  for p_tag in p_tags:
      file.write(p_tag.text.strip() + '\n')

    # Close the file
  file.close()

  
    
```
