```python
import pandas as pd
```

```python
df_dataset = pd.read_csv("/content/drive/MyDrive/file/D_02-21-2018.csv (1).csv")
```

```python
df_dataset = df_dataset.reset_index(drop=True)
```

```python
# Remove erroneous 'Label' row
df_dataset.drop(df_dataset.loc[df_dataset["Label"] == "Label"].index, inplace=True)
```

```python
# Forcing them to be strings
df_dataset = df_dataset.astype({"Protocol": str})
```

```python
df_dataset = pd.get_dummies(df_dataset, columns=['Protocol'], drop_first=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Compute the variance of each column
variances = df_dataset.var()
fig, ax = plt.subplots(figsize=(20, 6))
# Plot the variances as a bar chart
variances.plot(kind='bar')

# Show the plot
plt.show()
```
